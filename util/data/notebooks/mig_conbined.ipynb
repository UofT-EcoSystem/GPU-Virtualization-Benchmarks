```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import pickle
import os
import sys
from ipywidgets import widgets
from IPython.display import display, HTML
from IPython.display import clear_output
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('../pickles/mig_v100_combined.pkl')

df.drop(df.columns.difference(['pair_str','norm_ipc','sm_clusters']), 1, inplace=True)
df = df.round(3)
df
```

```python
# form a dictionary with mig results
mig_results = {}
for benchmark in df['pair_str'].unique():
    bench_df = df.loc[df['pair_str'] == benchmark]
    bench_df.drop(bench_df.columns.difference(['norm_ipc','sm_clusters']), 1, inplace=True)
    bench_df = bench_df.sort_values(by=['sm_clusters'])
    print(bench_df)
    ipc_list = bench_df['norm_ipc'].to_list()
    mig_results[benchmark] = ipc_list
    #print(bench_df.to_dict())
```

```python
print(mig_results)
```
